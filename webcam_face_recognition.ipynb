```python
import face_recognition
import cv2
import os
from datetime import datetime

# knownãƒ•ã‚©ãƒ«ãƒ€ã®ç”¨æ„
known_dir = "known"
os.makedirs(known_dir, exist_ok=True)

# ç™»éŒ²ç”»åƒã®èª­ã¿è¾¼ã¿ã¨ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰
known_face_encodings = []
known_face_names = []

print("ğŸ§  é¡”ç”»åƒã®ç™»éŒ²ä¸­...")
for filename in os.listdir(known_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        path = os.path.join(known_dir, filename)
        image = face_recognition.load_image_file(path)
        encoding = face_recognition.face_encodings(image)
        if encoding:
            known_face_encodings.append(encoding[0])
            name = os.path.splitext(filename)[0]
            known_face_names.append(name)
            print(f"âœ… {filename} ç™»éŒ²æˆåŠŸï¼ˆ{name}ï¼‰")
        else:
            print(f"âš ï¸ {filename} ã«é¡”ãŒæ¤œå‡ºã§ãã¾ã›ã‚“ã§ã—ãŸ")

if not known_face_encodings:
    print("âŒ ç™»éŒ²ã•ã‚ŒãŸé¡”ç”»åƒãŒã‚ã‚Šã¾ã›ã‚“ã€‚known/ ãƒ•ã‚©ãƒ«ãƒ€ã«è¿½åŠ ã—ã¦ãã ã•ã„ã€‚")
    exit()

# Webã‚«ãƒ¡ãƒ©èµ·å‹•
video_capture = cv2.VideoCapture(0)
print("ğŸ“¸ Webã‚«ãƒ¡ãƒ©èµ·å‹•ä¸­ï¼ˆPress 's'ã§ä¿å­˜ã€'q'ã§çµ‚äº†ï¼‰")

while True:
    ret, frame = video_capture.read()
    if not ret:
        print("âŒ ã‚«ãƒ¡ãƒ©ã‹ã‚‰æ˜ åƒã‚’å–å¾—ã§ãã¾ã›ã‚“ã§ã—ãŸ")
        break

    # BGR â†’ RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # é¡”æ¤œå‡ºã¨ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    print(f"ğŸ§· æ¤œå‡ºé¡”æ•°ï¼š{len(face_locations)}")

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # é¡ä¼¼åº¦ã‚¹ã‚³ã‚¢è¨ˆç®—
        face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)

        if len(face_distances) == 0:
            continue

        best_match_index = face_distances.argmin()
        confidence = 1 - face_distances[best_match_index]

        matches = face_recognition.compare_faces(known_face_encodings, face_encoding, tolerance=0.65)
        if matches[best_match_index]:
            name = f"{known_face_names[best_match_index]} ({confidence:.2f})"
        else:
            name = "Unknown"

        print(f"ğŸ“Œ åˆ¤å®šçµæœ: {name}")

        # æ ã¨ãƒ©ãƒ™ãƒ«æç”»
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(frame, name, (left, top - 10), cv2.FONT_HERSHEY_DUPLEX, 1.0, (0, 0, 0), 2)

    # è¡¨ç¤º
    cv2.imshow("Webcam Face Recognition", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        print("ğŸ›‘ çµ‚äº†ã—ã¾ã™")
        break
    elif key == ord('s'):
        now = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"captured_{now}.jpg"
        cv2.imwrite(filename, frame)
        print(f"ğŸ’¾ ç”»åƒä¿å­˜ï¼š{filename}")

# çµ‚äº†å‡¦ç†
video_capture.release()
cv2.destroyAllWindows()
```
